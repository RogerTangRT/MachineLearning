# Creating statistical Graphics in Seaborn

```python
# Imports
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
```

```python
%matplotlib inline
rcParams['figure.figsize']= 5,6
sns.set_style('whitegrid')
```

## Eyeballing database distributions with histograms

```python
# Preparando os dados
# Lê do arquivo csv
address = r'G:\Meu Drive\Cursos\Python\MachineLearning\cars.csv'
cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
cars.index = cars.car_names
cars.head()
```

```python
mpg = cars['mpg']
mpg.plot(kind='hist')
```

```python
plt.hist(mpg)
plt.plot()
```

```python
# Usando seaborn
sns.displot(mpg)
```

## Seeing scatterplots in action

```python
cars.plot(kind='scatter',x='hp', y='mpg', c=['darkgray'], s=150)
```

```python
## Usando seaborn
sns.regplot(x='hp', y='mpg', data=cars, scatter=True)
```

## Generating a scatter plot matrix

```python
sns.pairplot(cars)
```

```python
cars_subset = cars[['mpg', 'disp','hp','wt']]
sns.pairplot(cars_subset)
```

## Building Boxplot

```python
# mpg - Miles per Galon
# am - Automatic Transmission
cars.boxplot(column='mpg',by='am')
# wt - Weight 
# am - Automatic Transmission
cars.boxplot(column='wt',by='am')
```

```python
sns.boxplot(x='am',y='mpg', data=cars, hue='am', palette='hls')
```
