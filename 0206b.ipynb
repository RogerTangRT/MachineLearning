```python
# Importando as bibliotecas
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
```

## Concatenando Dados

```python
DF_obj = DataFrame(np.arange(36).reshape(6,6))
DF_obj
```

```python
DF_obj2 = DataFrame(np.arange(15).reshape(5,3))
DF_obj2
```

```python
# Apenda as colunas do Dataframe 2 no dataframe 1
pd.concat([DF_obj,DF_obj2],axis=1)
```

```python
#concatena do Dataset 2 depois do Dataset 1
pd.concat([DF_obj,DF_obj2])
```

## Transformando Dados

## Apagando Dados

```python
# Remove as linhas 0 e 2
DF_obj.drop([0,2])
```

```python
# Remove as colunas chamadas 0 e 2
DF_obj.drop([0,2], axis=1)
```

## Adicionando Dados

```python
series_obj = Series(np.arange(6))
series_obj.name = 'added_variable'
series_obj
```

```python
variable_added = DataFrame.join(DF_obj,series_obj)
variable_added
```

```python
#  ignore_index=False. Os indices serão duplicados
added_datatable = pd.concat([variable_added,variable_added], ignore_index=False)
added_datatable
```

```python
#  ignore_index=True. Os indices serão continuados
added_datatable = pd.concat([variable_added,variable_added], ignore_index=True)
added_datatable
```
