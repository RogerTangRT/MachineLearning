## Instance-based learning w/ k-Nearest Neighbor

```python
import numpy as np
import pandas as pd
import scipy

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sns

import urllib
```

```python
import sklearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn import neighbors, preprocessing

from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
#Now let's set our plotting parameters for the Jupyter notebook.
np.set_printoptions(precision=4, suppress=True) 
%matplotlib inline
rcParams['figure.figsize'] = 7, 4
sns.set_style("whitegrid")
```

```python
# Descrição das colunas do dataset 'cars' (mtcars)
# car_names : Nome do carro.
#             Ex.: Mazda RX4
# mpg       : Consumo de combustível em milhas por galão (Miles Per Gallon).
#             Ex.: 21.0
# cyl       : Número de cilindros do motor.
#             Ex.: 4, 6 ou 8
# disp      : Cilindrada (deslocamento do motor), em polegadas cúbicas (Cubic Inches).
#             Ex.: 160.0
# hp        : Potência do motor, em cavalos-vapor (Horsepower).
#             Ex.: 110
# drat      : Relação do eixo traseiro (Rear Axle Ratio).
#             Ex.: 3.90
# wt        : Peso do veículo, em milhares de libras.
#             Ex.: 2.620 = 2620 lb
# qsec      : Tempo para percorrer 1/4 de milha, em segundos.
#             Ex.: 16.46
# vs        : Tipo do motor.
#             0 = Motor em V
#             1 = Motor em linha (Straight)
# am        : Tipo de transmissão.
#             0 = Automática
#             1 = Manual
# gear      : Número de marchas da transmissão.
#             Ex.: 3, 4 ou 5
# carb      : Número de carburadores.
#             Ex.: 1, 2, 3, 4, 6 ou 8

address = 'G:\\Meu Drive\\Cursos\\Python\\MachineLearning\\Parte2\\data\\mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
cars
```

```python
#.values converte essa coluna em um array do NumPy.
# mpg       : Consumo de combustível em milhas por galão (Miles Per Gallon).
#             Ex.: 21.0
# disp      : Cilindrada (deslocamento do motor), em polegadas cúbicas (Cubic Inches).
#             Ex.: 160.0
# hp        : Potência do motor, em cavalos-vapor (Horsepower).
#             Ex.: 110
# wt        : Peso do veículo, em milhares de libras.
#             Ex.: 2.620 = 2620 lb

x_prime = cars[['mpg','disp','hp','wt',]].values
x_prime
```

```python
# y = cars.iloc[:, 9].values
# Explicação:
# .values Pega apenas os dados como array e não o dataset
# faz três coisas:
# cars é um objeto do tipo DataFrame da biblioteca pandas.
# .iloc[:, 9] seleciona dados usando índices numéricos:
# : significa todas as linhas.
# 9 significa a décima coluna (a contagem começa em 0). Coluna "am" (Tipo de transmissão.)
# Ou seja, isso equivale a:
# cars.iloc[:, 9]
# = "pegue todas as linhas da coluna de índice 9 "am" (Tipo de transmissão.)".
y = cars.iloc[:,9].values
y
```

```python
# A função scale() da biblioteca sklearn.preprocessing padroniza os dados,
# transformando cada coluna para que tenha:

# - Média = 0
# - Desvio padrão = 1

# A transformação aplicada a cada valor é:

#     x_padronizado = (x - μ) / σ

# onde:
#     x = valor original;
#     μ = média da coluna;
#     σ = desvio padrão da coluna.

# A padronização coloca todas as variáveis na mesma escala, evitando que
# atributos com valores maiores tenham influência desproporcional no modelo.
# Exemplo de variáveis com escalas diferentes:

# Variável        Faixa de valores
# --------------------------------
# hp (potência)  : 50 a 350
# wt (peso)      : 1.5 a 5.5
# gear (marchas) : 3 a 5
# A padronização evita que variáveis com escalas muito diferentes dominem o treinamento de um modelo.
# Sem padronização, a variável hp pode ter muito mais influência simplesmente porque seus números são maiores. 
# Após scale(), todas as variáveis ficam na mesma escala, permitindo que o algoritmo trate cada uma de forma 
# mais equilibrada.

x = preprocessing.scale(x_prime)
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=17)
```

```python
# Cria objeto KNeighborsClassifier
clf = neighbors.KNeighborsClassifier()
clf.fit(x_train,y_train)
```

```python
# y predito
y_pred = clf.predict(x_test)
# y esperado é o y de teste
y_expect = y_test

# cria metrica do modelo. Avaliando o y esperado com o y previsto
print(metrics.classification_report(y_expect,y_pred))
```
