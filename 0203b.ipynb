## Importa bibliotecas

```python
# Importando as bibliotecas
import numpy as np
import pandas as pd
from pandas import DataFrame
```

## Cria um dataset

```python
numbers_df = DataFrame(np.arange(0,90,3).reshape(10,3), index=['row 1','row 2','row 3','row 4','row 5','row 6','row 7','row 8','row 9','row 10'],columns = ['column 1','column 2','column 3'])
```

```python
# Mostra o Dataset
numbers_df
```

## Indexação Simples

```python
# Mostra o tipo e valor do elemento
# Linha 1 (0), coluna 2 (1)
numbers_df.iloc[0,1]
```

```python
# Alterando valor do elemento
numbers_df.iloc[0,1] = 20
numbers_df
```

## Indexação do Array

```python
# Primeira lista são as linhas e segunda as colunas
# Seleciona as linhas 2(1),3(2),5(4) das colunas 2(1), 3(2)
numbers_df.iloc[[1,2,4],[1,2]]
```

## Indexação Booleana
### Operadores de Comparação (> < = <= => == !=>) e mascaras

```python
# True os elementos que atendem a condição > 30
mask = numbers_df > 30
mask
```

```python
#retorna o Dataframe onde os valores da mascara é true. Os demais vaores são NaN (Not a Number)
numbers_df[mask]
```

```python
#Altera os valores do filtro > 30 para 0 
numbers_df[numbers_df > 30] = 0
numbers_df
```

## Dividir dados em partes (fatias)

```python
# Seleciona as linhas 3(2) a 6 e colunas 2(1) a 3
numbers_df.iloc[2:6,1:3]
```
