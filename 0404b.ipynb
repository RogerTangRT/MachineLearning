## Defining plor color

```python
x = range(1,10)
y = [1,2,3,4,0.5,4,3,2,1]

plt.bar(x,y)
```

```python
wide = [0.5,0.5,0.5,0.9,0.9,0.9,0.5,0.5,0.5]
color = ['salmon']

plt.bar(x,y,width=wide, color=color, align='center')
```

```python
address = r'.\cars.csv'
cars = pd.read_csv(address)
cars.columns = ['car_names','MilesPerGalon','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
mpg = cars['MilesPerGalon']
cars.head()
```

```python
z = [1,2,3,4,0.5]
plt.pie(z)
plt.show()
```

```python
color_theme = ['#A9A9A9','#FFA07A','#B0E0E6','#FFE4C4','#BDB76B']
plt.pie(z,colors=color_theme)
```

## Customizing line styles

```python
x1 = range(1,10)
y1 = [9,8,7,6,5,4,3,2,1]
plt.plot(x,y)
plt.plot(x1,y1)
```

```python
plt.plot(x,y, ls='solid', lw=5)
plt.plot(x1,y1, ls='--',lw=10)
```

## Set plot markers

```python
plt.plot(x,y,marker = '1',mew=20)
plt.plot(x1,y1, marker='+', mew=15)
```
