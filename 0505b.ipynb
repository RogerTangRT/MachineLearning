## The Spearman Rank Correlation

[Descrição dos Campos cars](Cars_DescriçãoCampos.md)

```python
address = r'G:\Meu Drive\Cursos\Python\MachineLearning\cars.csv'
cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
cars.head()
```

```python
sns.pairplot(cars)
```

```python
# Essas variáveis não são lineares. 
# As variáveis são ordinárias: Se são numéricas podem ser categorizadas em grupos.
# Cada variável assume apenas valores únicos possíveis
x = cars[['cyl','vs','am','gear']]
sns.pairplot(x)
```

```python
cyl = cars['cyl']
vs = cars['vs']
am = cars['am']
gear = cars['gear']

speranmanr_coefficient , p_value = spearmanr(cyl,vs)
print('Spearman Rank Correlation Coefficient %0.3f' % (speranmanr_coefficient))
```

```python
speranmanr_coefficient , p_value = spearmanr(cyl,am)
print('Spearman Rank Correlation Coefficient %0.3f' % (speranmanr_coefficient))
```

```python
speranmanr_coefficient , p_value = spearmanr(cyl,gear)
print('Spearman Rank Correlation Coefficient %0.3f' % (speranmanr_coefficient))
```

## Chi-square test for independence

```python
table = pd.crosstab(cyl,am)
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(table.values)
print('Chi-square Statistic %0.3f p_value %0.3f' % (chi2,p))
```

```python
table = pd.crosstab(cyl,vs)

chi2, p, dof, expected = chi2_contingency(table.values)
print('Chi-square Statistic %0.3f p_value %0.3f' % (chi2,p))
```

```python
table = pd.crosstab(cyl,gear)

chi2, p, dof, expected = chi2_contingency(table.values)
print('Chi-square Statistic %0.3f p_value %0.3f' % (chi2,p))
```
