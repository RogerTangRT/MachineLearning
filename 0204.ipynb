## Tratando valores faltantes
Dados faltantes podem ser substituídos por:

- Média dos dados das linhas. Na maioria dos casos é melhor preencher os dados faltantes por dados aproximados do que deixá-los como 0.

```python
# Importando as bibliotecas
import numpy as np
import pandas as pd
from pandas import DataFrame
```

## Preencher os dados faltantes usando fillna(), replace() e interpolate

```python
data = {'names':['steve','john','richard','sarah','randy','micheal','julie'],
        'age':[20,22,20,21,24,23,22],
        'gender':['Male','Male','Male','Female','Male','Male','Female'],
        'rank':[2,1,4,5,3,7,6]}
ranking_df = DataFrame(data)
ranking_df
```

```python
# Inserindo valores Nan
# Linhas 2:5 for (i=2;i<5;i++)
ranking_df.iloc[2:5,1] = np.nan
# Linhas 3:6 for (i=3;i<6;i++)
ranking_df.iloc[3:6,3] = np.nan
# ':' - Indica todas as colunas
ranking_df.iloc[3,:] = np.nan
ranking_df
ranking_df_original = ranking_df.copy()
```

## Descobrir o que está faltando (isnull() notnull())

```python
# Filtra dados faltantes
ranking_df.isnull()
```

```python
ranking_df.notnull()
```

```python
# Mostra dados faltantes em uma coluna específica.
bool_series = pd.isnull(ranking_df['age'])
ranking_df[bool_series]
```

```python
# Preenche com zeros o que está com NaN
ranking_df.fillna(0)
```

```python
# Preenchendo os dados faltantes NaN com o último valor válido encontrado
# A linha 3 que estava toda NaN também será preenchida com o valor da linha anterior válida
# ranking_df.fillna(method='pad')
ranking_df.ffill()
```

```python
# Preencher Dados Faltantes com backfill (bfill) . Dó último valor válido para cima
# Agora a linha 3 foi preenchida com o valor da próxima linha válida
# ranking_df.fillna(method='bfill')
ranking_df.bfill()
```

```python
# Preenchendo com valores interpolados linearmente
# ranking_df.interpolate(method='linear')

# A linha acima gera erro pois tenta interpolar os textos.

# Interpola apenas as coluna numéricas
# ranking_df = ranking_df.copy()

numeric_cols = ranking_df.select_dtypes(include=['number']).columns
ranking_df[numeric_cols] = ranking_df[numeric_cols].interpolate(method='linear')
ranking_df
```

```python
ranking_df_original
```

```python
# Apaga tudo que tem NaN 
ranking_df_original.dropna()
```

```python
# Apaga apenas as colunas onde todos os valores são NaN
ranking_df_original.dropna(how='all')
```

```python
# Exclui todas as colunas que possuem pelo menos um valor NaN.
# ranking_df_original, possui a linha 3 com todos valores NaN
ranking_df_original.dropna(axis=1)
```
