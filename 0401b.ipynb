```python
# Importando bibliotecas
import matplotlib.pyplot as plt
import seaborn as sns

from pandas import DataFrame
```

## Criando Dataset

```python
data = {'names':['steve','john','richard','sarah','randy','micheal','julie'],
        'age':[20,22,20,21,24,23,22],
        'gender':['Male','Male','Male','Female','Male','Male','Female'],
        'rank':[2,1,4,5,3,7,6]}
df = DataFrame(data)
df
```

## Grafico de Barras do Matplotlib

```python
# Criando grafico de Barras com coluna do nome
plt.bar(df['names'],df['age'])
# Label do X
plt.xlabel('Names')
# Label do Y
plt.ylabel('Age')
plt.title('Comparing Ages')
plt.show()
```

## Grafico de Barras do Seaborn

```python
plot = sns.barplot(data=df, x='names',y='age')
plot.set_title("Comparing Ages")
```

## Gráfico de Linha Mathplotlib

```python
plt.plot(df['names'], df['age'])
# Label do X
plt.xlabel('Names')
# Label do Y
plt.ylabel('Age')
plt.title('Comparing Ages')
plt.show()
```

## Gráfico de Linha Seaborn

```python
plot = sns.lineplot(data=df,x='names',y='age')
plt.show()
```

## Gráfico de Pizza Matplotlib

```python
plt.pie(df['age'],labels=df['names'])
plt.title("Age Comparison")
plt.show()
```

## Grafico de Pizza Seaborn

```python
colors = sns.color_palette('pastel')[0:5]
plt.pie(df['age'],labels=df['names'],colors=colors)
plt.show()
```
